```python
import networkx as nx
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# read files
# Cook
flow_20_march_week4_cook = gpd.read_file('../../data/processed_data/flow_20_march_week4_Cook_shp.geojson')
flow_16_acs_cook = gpd.read_file('../../data/processed_data/flow_16_acs_Cook_shp.geojson')
cook_tracts = gpd.read_file('../../data/processed_data/us_tract_Cook_shp.geojson')
# MA
flow_20_march_week4_ma = gpd.read_file('../../data/processed_data/flow_20_march_week4_MA_shp.geojson')
flow_16_acs_ma = gpd.read_file('../../data/processed_data/flow_16_acs_MA_shp.geojson')
ma_tracts = gpd.read_file('../../data/processed_data/us_tract_MA_shp.geojson')
```

```python
# 
print(cook_tracts.shape)
print(flow_20_march_week4_cook.shape)
print(flow_16_acs_cook.shape)
```

```python
# create spatial network
from pysal.lib import weights
cook_tract_net = weights.contiguity.Queen.from_dataframe(cook_tracts)
ma_tract_net = weights.contiguity.Queen.from_dataframe(ma_tracts)

cook_tract_net=cook_tract_net.to_networkx()
ma_tract_net=ma_tract_net.to_networkxworkx()
```

```python
# degree distributions for cook spatial, ma spatial, cook flow, ma flow, etc.
import collections

# cook tracts
degree_sequence = sorted([d for n, d in cook_tract_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')

# ma tracts
degree_sequence = sorted([d for n, d in ma_tract_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')
```

```python
# degree distribution
# cook flow (discrete only)
flow_20_cook_net = nx.from_pandas_edgelist(flow_20_march_week4_cook, 'geoid_o', 'geoid_d', edge_attr='pop_flows')
degree_sequence = sorted([d for n, d in flow_20_cook_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')

flow_16_cook_net = nx.from_pandas_edgelist(flow_16_acs_cook, 'Ofips', 'Dfips', edge_attr='Flow')
degree_sequence = sorted([d for n, d in flow_16_cook_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')

# ma flow (discrete only)
flow_20_ma_net = nx.from_pandas_edgelist(flow_20_march_week4_ma, 'geoid_o', 'geoid_d', edge_attr='pop_flows')
degree_sequence = sorted([d for n, d in flow_20_ma_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')

flow_16_ma_net = nx.from_pandas_edgelist(flow_16_acs_ma, 'Ofips', 'Dfips', edge_attr='Flow')
degree_sequence = sorted([d for n, d in flow_16_ma_net.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(deg, cnt, width = 0.8, color='b')
```

```python
# number of nodes and edges
# cook_tract_net, ma_tract_net, flow_20
print(len(cook_tract_net.nodes))
print(len(ma_tract_net.nodes)) # ma tracts are not connected.
print(len(flow_20_cook_net.nodes))
print(len(flow_20_ma_net.nodes))
print(len(flow_16_cook_net.nodes))
print(len(flow_16_ma_net.nodes))

print(np.log10(len(cook_tract_net.nodes)))
print(np.log10(len(ma_tract_net.nodes))) # ma tracts are not connected.
print(np.log10(len(flow_20_cook_net.nodes)))
print(np.log10(len(flow_20_ma_net.nodes)))
print(np.log10(len(flow_16_cook_net.nodes)))
print(np.log10(len(flow_16_ma_net.nodes)))
```

```python
# visual
plt.figure(figsize=(8,8))
pos = nx.spring_layout(cook_tract_net)
nx.draw_networkx_nodes(cook_tract_net, pos, node_size = 10)
nx.draw_networkx_edges(cook_tract_net, pos, alpha=0.2)

# sw: what are the good network visualization tools?
```

```python
# check connectivity
print(nx.is_connected(cook_tract_net))
print(nx.is_connected(ma_tract_net)) # ma tracts are not connected.
print(nx.is_connected(flow_20_cook_net))
print(nx.is_connected(flow_20_ma_net))
print(nx.is_connected(flow_16_cook_net))
print(nx.is_connected(flow_16_ma_net))
```

```python
# check path length (diameter)
print("Cook census tracts' diameter is: ", nx.diameter(cook_tract_net))
# print("MA census tracts' diameter is: ", nx.diameter(ma_tract_net))
print("Cook flow 2020 diameter is: ", nx.diameter(flow_20_cook_net))
print("MA flow 2020 diameter is: ", nx.diameter(flow_20_ma_net))
print("Cook flow 2016 diameter is: ", nx.diameter(flow_16_cook_net))
print("MA flow 2016 diameter is: ", nx.diameter(flow_16_ma_net))
```
